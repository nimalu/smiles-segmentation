```python
from ultralytics import YOLO
import glob
```

```python
model = YOLO("yolo11n-seg-smiles-best.pt")
```

```python
images = glob.glob("../smiles-rendering/dataset/images/val/*.png")[:5]

results = model(images) 

result_images = []
for result in results:
    result_images.append(result.plot(pil=True, font_size=10, color_mode="instance", line_width=1))
```
